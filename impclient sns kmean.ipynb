```python
# now build the plot 
# run a for loop from k =1 , till k = 10
```

```python
ssd = []

for k in range(1,11):
    kmean_sns = KMeans(n_clusters=k)
    kmean_sns.fit(sns)
    score1 = kmean_sns.score(sns)
    ssd.append(score1)
```

```python
import numpy as np # import library
```

```python
import matplotlib.pyplot as plt 
```

```python
ssd = np.abs(ssd)
```

```python
ssd = np.round(ssd)
```

```python
plt.plot(ssd)
```
